```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def data_preprocessing(data):
    data["timestep_diff"]=data["timestep"] - data["timestep"].shift(1).fillna(0)
    data["average_reward"] = data["reward"] / data["timestep_diff"]
    return data
```

```python
AV_num = 75
AV_length = 300
AV1 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_1.csv",index_col=None)
AV2 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_2.csv",index_col=None)
AV3 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_3.csv",index_col=None)
AV4 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_4.csv",index_col=None)
AV5 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_5.csv",index_col=None)
AV1 = data_preprocessing(AV1)
AV2 = data_preprocessing(AV2)
AV3 = data_preprocessing(AV3)
AV4 = data_preprocessing(AV4)
AV5 = data_preprocessing(AV5)



AV_1 = pd.concat([AV1[:AV_length],AV2[:AV_length],AV3[:AV_length],AV4[:AV_length],AV5[:AV_length]])
```

```python
AV_num = 70
#AV_length = 300
AV1 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_1.csv",index_col=None)
AV2 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_2.csv",index_col=None)
AV3 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_3.csv",index_col=None)
AV4 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_4.csv",index_col=None)
AV5 = pd.read_csv("runs/sumo-v0/DQN_sumo-v0_log_"+str(AV_num)+"_5.csv",index_col=None)
AV1 = data_preprocessing(AV1)
AV2 = data_preprocessing(AV2)
AV3 = data_preprocessing(AV3)
AV4 = data_preprocessing(AV4)
AV5 = data_preprocessing(AV5)
AV_2 = pd.concat([AV1[:AV_length],AV2[:AV_length],AV3[:AV_length],AV4[:AV_length],AV5[:AV_length]])
```

```python
AV = pd.concat([AV_1,AV_2])
```
