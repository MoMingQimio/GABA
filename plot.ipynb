```python
def data_preprocessing(data):
    data["timestep_diff"]=data["timestep"] - data["timestep"].shift(1).fillna(0)
    data["average_reward"] = data["reward"] / data["timestep_diff"]
    return data
```

```python
AV_num = 75
AV_length = 300
AV1 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_1.csv",index_col=None)
AV2 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_2.csv",index_col=None)
AV3 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_3.csv",index_col=None)
AV4 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_4.csv",index_col=None)
AV5 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_5.csv",index_col=None)
AV1 = data_preprocessing(AV1)
AV2 = data_preprocessing(AV2)
AV3 = data_preprocessing(AV3)
AV4 = data_preprocessing(AV4)
AV5 = data_preprocessing(AV5)



AV = pd.concat([AV1[:AV_length],AV2[:AV_length],AV3[:AV_length],AV4[:AV_length],AV5[:AV_length]])
```

```python
def BV_data_preprocessing(data):
    data["adversarial_counts_diff"]=data["adversarial_counts"] - data["adversarial_counts"].shift(1).fillna(0)
    data["average_reward"] = data["reward"] / (data["adversarial_counts_diff"]+1)
    return data
```

```python
BV_num = 44
length = 100
BV1 = pd.read_csv("PPO_logs\sumo-v0\PPO_sumo-v0_log_"+str(BV_num)+"_1.csv",index_col=None)
BV2 = pd.read_csv("PPO_logs\sumo-v0\PPO_sumo-v0_log_"+str(BV_num)+"_2.csv",index_col=None)
BV3 = pd.read_csv("PPO_logs\sumo-v0\PPO_sumo-v0_log_"+str(BV_num)+"_3.csv",index_col=None)
BV4 = pd.read_csv("PPO_logs\sumo-v0\PPO_sumo-v0_log_"+str(BV_num)+"_4.csv",index_col=None)
BV5 = pd.read_csv("PPO_logs\sumo-v0\PPO_sumo-v0_log_"+str(BV_num)+"_5.csv",index_col=None)
BV1 = BV_data_preprocessing(BV1)
BV2 = BV_data_preprocessing(BV2)
BV3 = BV_data_preprocessing(BV3)
BV4 = BV_data_preprocessing(BV4)
BV5 = BV_data_preprocessing(BV5)



BV = pd.concat([BV1[:length],BV2[:length],BV3[:length],BV4[:length],BV5[:length]])
```
