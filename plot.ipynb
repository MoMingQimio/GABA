```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
AV_num = 32
AV1 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_1.csv",index_col=None)
AV2 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_2.csv",index_col=None)
AV3 = pd.read_csv("runs\sumo-v0\DQN_sumo-v0_log_"+str(AV_num)+"_3.csv",index_col=None)
AV = pd.concat([AV1[:300],AV2[:300],AV3[:300]])
```

```python
sns.lineplot(x=AV["episode"],y=AV["collision_rate"])
```

```python
sns.lineplot(x=AV["episode"],y=AV["reward"])
```
